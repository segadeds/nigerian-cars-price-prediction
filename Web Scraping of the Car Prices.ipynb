```python
import requests
from bs4 import BeautifulSoup
import pandas
```

```python
z = 1
l = []
stem = 'https://cars45.com'
for page in range(1,274):
    url = "https://cars45.com/listing?price_min=&price_min=999999999"
    if z == 1:
        r = requests.get(url)
    else:
        r = requests.get('https://cars45.com/listing'+'/page'+str(page)+'?price_max=999999999&listing_id=ODp8Yi8p4uAU7Nxg')
    z = z + 1
    c = r.content
    soup = BeautifulSoup(c,'html.parser')
    all = soup.find_all('a', {'class':"car-feature"})
    for i in range(0,len(all)):
        link = all[i].get('href')
        new_url = stem+link 
        #print(new_url)
        r_2 = requests.get(new_url)
        c_2 = r_2.content
        soup = BeautifulSoup(c_2,'html.parser')
        info = soup.find('div', {'class':"general-info grid"}).find_all('div') 
        d = {}
        for i in info:
            if i.find('span').text == 'Make':
                d['Make'] = i.find('p').text
            if i.find('span').text == 'Year of manufacture':
                d['Year of manufacture'] = i.find('p').text
            if i.find('span').text == 'Condition':
                d['Condition'] = i.find('p').text
            if i.find('span').text == 'Mileage':
                d['Mileage'] = i.find('p').text
            if i.find('span').text == 'Engine Size':
                d['Engine Size'] = i.find('p').text
        extra = soup.find('div', {'class':"svg flex"}).find_all('div')
        for i in extra:
            if i.find('span').text == 'SUV':
                d['Build'] = i.find('span').text
            if i.find('span').text == 'Petrol':
                d['Fuel'] = i.find('span').text
            if i.find('span').text == 'Hybrid':
                d['Fuel'] = i.find('span').text
            if i.find('span').text == 'Electric':
                d['Fuel'] = i.find('span').text
            if i.find('span').text == 'Diesel':
                d['Fuel'] = i.find('span').text
            if i.find('span').text == 'AMT':
                d['Transmission'] = i.find('span').text
            if i.find('span').text == 'Automatic':
                d['Transmission'] = i.find('span').text
            if i.find('span').text == 'CVT':
                d['Transmission'] = i.find('span').text
            if i.find('span').text == 'Manual':
                d['Transmission'] = i.find('span').text
        d['Price'] = soup.find('div', {'class':"main-details__name"}).find('h5').text.replace('â‚¦ ','')
        #print(d)
        l.append(d)
    print(page)
```

```python
df = pandas.DataFrame(l)
df
```

```python
df.to_csv('Nigerian_Car_Prices.csv')
```
